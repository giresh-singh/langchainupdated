## Model Integration With OpenAI, Google Gemini and Groq 

```python
### Load environment variables from .env file and set API keys for OpenAI, Groq, and Google
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
### import init_chat_model from langchain to initialize chat models
from langchain.chat_models import init_chat_model
```

#### OpenAI Chat Model Integration

```python
#### load OpenAI Chat model Integration
model = init_chat_model(
    model="gpt-4.1-mini",
    model_provider="openai"
    )  # OpenAI model
```

```python
### run the agent and take response in response variable
response = model.invoke("Hello! How are you?")
print(response.content)
```

#### Google Gemini Chat Model Integration

```python
### Google Gemini Chat model Integration
model = init_chat_model(
    model="gemini-2.5-flash-lite",
    model_provider="google_genai"
)  # Google Gemini model
```

```python
response = model.invoke("Explain Python.")

print(response.content)
```

#### Groq Chat Model Integration

```python
model = init_chat_model(
    model="llama-3.1-8b-instant",
    model_provider="groq"
)  # Groq model
```

```python
response = model.invoke("What is LangChain?")
print(response.content)
```

#### Invoke, Streaming and Batch methods in LangChain

##### Invoke - User get output at once and till then need to wait
##### Streaming - Gives output in chunks need not to wait longer
##### Batch - is used to process multiple inputs in single call

```python

for response in model.stream("What is LangChain? and LangGraph?"):
    print(response.content,end="",flush=True)
```

```python
questions = [
    "What is LangChain?",
    "What is LangGraph?", 
    "What is the difference between LangChain and LangGraph?"
    ]
for response in model.batch(questions):
    print(response.content)
```

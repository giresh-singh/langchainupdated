```python
### Load environment variables from .env file and set API keys for OpenAI, Groq, and Google
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
### import init_chat_model from langchain to initialize chat models
from langchain.chat_models import init_chat_model
## import AIMessage, HumanMessage, SystemMessage from langchain.messages to create messages for the chat model
from langchain.messages import AIMessage, HumanMessage, SystemMessage
```

```python
### Groq Llama Chat model Integration
model = init_chat_model(
    model="llama-3.1-8b-instant",
    model_provider="groq"
)  # Groq model
```

```python

messages = [
    SystemMessage(content="You are a poetry expert."),
    HumanMessage(content="write a poem about the ocean"),
]

response = model.invoke(messages)
response.content
```

```python
system_msg = SystemMessage(content="You are a helpful assistant.")
messages = [
    system_msg,
    HumanMessage(content="How do I create REST API?"),
]
resposnse = model.invoke(messages)
print(resposnse.content)
```

```python
## Detailed System messages and Human messages for better context and understanding of the conversation
system_msg = SystemMessage(""" you are a senior Python developer with experties in Web frameworks.
                           Always provide detailed and well explained answers with code examples and best practices
                           Be consise and clear in your answers. 
                           """)
messages = [
    system_msg,
    HumanMessage(content="How do I create REST API in Python?"),
]
resposnse = model.invoke(messages)
print(resposnse.content)
```

```python
## Adding meta data to the messages for better context and understanding of the conversation
human_msg = HumanMessage(
    content="Hello", ## optional: Identifier for different users in a multi-user conversation
    name="Giresh Singh", ## Optional: Unique identifier for the message, useful for tracking and referencing specific messages in a conversation
    id="msg-001", ## Optional: Additional metadata for the message, can be used to store any extra information relevant to the message
)
```

```python
response = model.invoke([
    human_msg])
response
```

```python
## create an AI message manually (e.g. for conversation history)
ai_msg = AIMessage(content="I would be happy to help you with that question.")
## Add to conversation history
messages=[
    SystemMessage("You are a helpful assistant."),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great ! What is 2 plus 2 ?")
]

response = model.invoke(messages)
print(response.content)
```
